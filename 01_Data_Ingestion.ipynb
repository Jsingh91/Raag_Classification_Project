```python
file_path = "/content/drive/MyDrive/Raag_Classifier/data/raw/SikhJS-0.7.0.zip"
with zipfile.ZipFile(file_path,'r') as my_zip_tool:
  list_of_files = my_zip_tool.namelist()
  print(len(list_of_files))
```

```python

ang_files = [file for file in list_of_files if file.startswith("SikhJS-0.7.0/assets/docs/json/SGGS")]
print(len(ang_files))
```

```python
all_ang_files = [file for file in ang_files if "Ang " in file]
```

```python
len(all_ang_files)
```

```python
all_ang_files[:10]
```

```python
def process_ang(filename,zip_tool):
  with zip_tool.open(filename) as file:
      data = json.load(file)
      return data
```

```python
ggs_data = []
with zipfile.ZipFile(file_path,'r') as my_zip_tool:
  for filename in all_ang_files:
    page_content = process_ang(filename,my_zip_tool)
    ggs_data.append(page_content)
```

```python
print(len(ggs_data))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```
