```python
pdf = pdfplumber.open("Siri Guru Granth Sahib in Gurmukhi, with Index (Uni).pdf")
```

```python
gurbani_full_text_pages = pdf.pages[:1483]
gurbani_full_text = [page.extract_text() for page in gurbani_full_text_pages]
```

```python
all_text = " ".join(gurbani_full_text)
```

```python
total_count = all_text.count("ਪ੍ਰਸਾਤਿ")
```

```python
print(total_count)
```

```python
total_count_2 = all_text.count("ਪ੍ਰਸਾਦਿ")
print(total_count_2)
```

```python
mahalla_count = all_text.count("ਮਿਲਾ")
print(mahalla_count)
```

```python
# Find the position of the first few occurrences of "ਮਿਲਾ" and print some context around them
position = all_text.find("ਮਿਲਾ",240000)
print(position)
all_text[position-2000:position+1000]
```

```python
import re

gurmukhi_pattern = r"([\u0A00-\u0A7F]+)\s+ਮਿਲਾ\s+([\u0A00-\u0A7F]+)"
matches = re.findall(gurmukhi_pattern,all_text)
print(f"Found {len(matches)} matches")
print(matches[:10])
```

```python
filtered_matches = [match for match in matches if not match[0].isdigit()]
```
