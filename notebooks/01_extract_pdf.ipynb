## Use Regex to extract Raag names and page numbers

```python
import re


# Test with an example line
test_line = "❁ ੧ਓ ਸਤਿਗੁਰ ਪ੍ਰਸਾਤਿ ॥ ਰਾਗੁ ਟੋਡੀ ............................... ੭੧੧ ਰਾਗੁ ਪ੍ਰਭਾਿੀ ........................ ੧੩੨੭"
print("Analyzing the pattern")
print(f"Original: {test_line}")
```

```python
# apply regex for patterns
pattern1 = r"ਰਾਗੁ\s+([^\.]+)\s*\.+\s*([੦-੯]+)"

matches = re.findall(pattern1,test_line)
print(f"Matches found: {matches}")
```

```python
import pandas as pd

def extract_raag_mapping(index_text):
    # Clean the text(remove decorative symbols)
    cleaned_text = re.sub(r"❁", "", index_text) # Remove flowers
    
    # regex pattern
    pattern = r"ਰਾਗੁ\s+([^\.]+)\s*\.+\s*([੦-੯]+)"

    matches = re.findall(pattern, cleaned_text)

    # Create a Dataframe
    raag_df = pd.DataFrame(matches, columns=['Raag_Name', 'Start_page_Gurmukhi'])

    print(f"Found {len(raag_df)} Raag entries!")
    return raag_df
```

```python
# Test the function
raag_map = extract_raag_mapping(index_text)
print(raag_map.head(30))
```

```python
# Known 31 main Raags of Guru Granth Sahib
known_raags = [
    "ਸਿਰੀ", "ਮਾਝ", "ਗਉੜੀ", "ਆਸਾ", "ਗੂਜਰੀ", "ਦੇਵਗੰਧਾਰੀ", "ਬਿਹਾਗੜਾ", 
    "ਵਡਹੰਸ", "ਸੋਰਠਿ", "ਧਨਾਸਰੀ", "ਜੈਤਸਰੀ", "ਟੋਡੀ", "ਬੈਰਾਡੀ", "ਤਿਲੰਗ", 
    "ਸੂਹੀ", "ਬਿਲਾਵਲ", "ਗੋਂਡ", "ਰਾਮਕਲੀ", "ਨਟ ਨਾਰਾਇਨ", "ਮਾਲੀ ਗਉੜਾ", "ਮਾਰੂ", 
    "ਤੁਖਾਰੀ", "ਕੇਦਾਰਾ", "ਭੈਰਉ", "ਬਸੰਤ", "ਸਾਰੰਗ", "ਮਲਾਰ", "ਕਾਨੜਾ", 
    "ਕਲਿਆਨ", "ਪ੍ਰਭਾਤੀ", "ਜੈਜਾਵੰਤੀ"
]

# Let's find which ones are missing from our extraction
def find_mising_raags(extracted_raags, known_raags):
    extracted_set = set(extracted_raags)
    known_set = set(known_raags)

    missing = known_set - extracted_set
    extra = extracted_set - known_set

    print(f"Missing Raags: {missing}")
    print(f'Extra/Unrecognized: {extra}')
    return missing,extra
```

```python
# Run the above function
missing,extra = find_mising_raags(raag_map["Raag_Name"].tolist(), known_raags)
```

## Handle Spelling Variations with Fuzzy Matching
Since we have character recogmition issues, let's use fuzzy matching.

```python
from fuzzywuzzy import process

def corrected_raag_names(extracted_raags, known_raags):
    corrected = []
    corrections = {}

    for raag in extracted_raags:
        # Find the best match from known Raags
        match, score = process.extractOne(raag, known_raags)
        if score > 70: # If good enough match
            corrected.append(match)
            corrections[raag] = match
        else:
            corrected.append(raag) # Keep original if no good match
    print("Spelling corrections needed:")
    for original, corrected in corrections.items():
        print(f" '{original}' -> '{corrected}'")
    return corrected, corrections
        
```

```python
correct_raag_names = corrected_raag_names(raag_map["Raag_Name"], known_raags)
```
