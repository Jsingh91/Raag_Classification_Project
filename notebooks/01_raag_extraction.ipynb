```python
import pdfplumber

def get_page_text(pdf_path, page_number=0):
    # open the pdf
    try:
         with pdfplumber.open(pdf_path) as pdf:
            page = pdf.pages[page_number]
            # Extract text from that page
            text = page.extract_text()
            if text:
                print(f'Text extracted from page {page_number + 1}!')
            else:
                print(f"No text found on page {page_number + 1}.")
                return None
            return text
    except FileNotFoundError:
        print(f'Cannot find the file', pdf_path)
    except Exception as e:
        print(f"Something went wrong {e}")
pdf_path = "../data/raw_text/Gurbani_with_index_Uni.pdf"
```

```python
import pdfplumber

def get_full_text(pdf_path):
    all_pages = []
    try:
         with pdfplumber.open(pdf_path) as pdf:
            for page_number, page in enumerate(pdf.pages):
                text = page.extract_text()
            
                if text:
                    all_pages.append(text)
                else:
                    print(f"No text found on page {page_number + 1}.")
                                  
    except FileNotFoundError:
        print(f'Cannot find the file', pdf_path)
    except Exception as e:
        print(f"Something went wrong {e}")
    full_text = "\n\n".join(all_pages)
    return full_text
pdf_path = "../data/raw_text/Gurbani_with_index_Uni.pdf"
```

```python
raw_gurbani_text = get_full_text(pdf_path)
len(raw_gurbani_text)
raw_gurbani_text[:500]
```

```python
def clean_gurbani_text(raw_text):
    # Normalize line breaks
    text = raw_text.replace('\r\n', '\n').replace('\r', '\n')

    # 2. Remove zero-width/invisible junk
    pattern_zero_width = r"[\ufeff\u200b\u200c\u200d\u200e\u200f]"
    text = re.sub(pattern_zero_width, " ", text)

    # 3. Remove characters that are NOT (newline, word chars, whitespace, Gurmukhi)
    pattern_garbage = r"[^\n\w\s\u0A00-\u0A7F]+"
    text = re.sub(pattern_garbage, " ", text)

    # 4. Tidy whitespace
    text = re.sub(r"[ \t]+", " ", text)
    text = re.sub(r"\n{3,}", "\n\n", text)

    return text.strip()
```

```python
cleaned_gurbani_text = clean_gurbani_text(raw_gurbani_text)
```
